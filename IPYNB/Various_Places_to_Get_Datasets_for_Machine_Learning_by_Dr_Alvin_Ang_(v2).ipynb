https://archive.ics.uci.edu/dataset/15/breast+cancer+wisconsin+original

```python
import pandas as pd

df = pd.read_csv('/content/breast-cancer-wisconsin.data',
                  names = [
                         '1. Sample code number',
                         '2. Clump Thickness (1 - 10)',
                         '3. Uniformity of Cell Size (1 - 10)',
                         '4. Uniformity of Cell Shape (1 - 10)',
                         '5. Marginal Adhesion (1 - 10)',
                         '6. Single Epithelial Cell Size (1 - 10)',
                         '7. Bare Nuclei (1 - 10)',
                         '8. Bland Chromatin (1 - 10)',
                         '9. Normal Nucleoli (1 - 10)',
                         '10. Mitoses (1 - 10)',
                         '11. Class: (2 for benign, 4 for malignant)'
                         ])
```

```python
df
```

##1b) Renaming Columns

```python
df.rename(columns = {"11. Class: (2 for benign, 4 for malignant)":"Class"}, inplace = True)

#rename the "Class" column
```

```python
df["Class"] = df["Class"].map({2:0, 4:1})

#0: benign (no cancer)
#1: malignant (cancer)
```

##6b) Import Dataset

```python
from sklearn.datasets import fetch_california_housing

ch = fetch_california_housing(as_frame=True)
print(ch.DESCR)
```

##6c) Print Neat Dictionary

```python
def print_neat_dictionary(x):
  for key, value in x.items():
    print(f"\n{key.upper()}:\n\n{value}")

print_neat_dictionary(ch)
```
